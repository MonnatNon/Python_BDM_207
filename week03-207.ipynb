```python
#Vector = 1D Array
a = np.array([1,2,3,4,5,6,7])
a
```

```python
#เลือกเอาตำเเหน่ง 2-5 ใน array มา output
a[2:5]
```

```python
#เลือกเอา 6 ตำเเหน่งมาจาก array แล้วมา output
a[:6]
```

```python
#เลือกเอาตำเเหน่งตั้งแต่ 3 จนตำแหน่งสุดท้ายมาจาก array แล้วมา output
a[3:]
```

```python
#เลือกเอาแค่ตำเเหน่งสุดท้ายมาจาก array แล้วมา output โดยใช้การ Negative number (-x) หรือ นับจากข้างหลังมาข้างหน้า
a[-1]
```

```python
#เลือกเอาแค่ตำเเหน่งรองสุดท้ายมาจาก array แล้วมา output โดยใช้การ Negative number (-x) หรือ นับจากข้างหลังมาข้างหน้า
a[-2]
```

```python
#เลือกเอาแค่ตำเเหน่ง 2 (positive number) จนไปถึง -3 (negative number) #ย้ำต้อง ตน. เป็น Array ที่ใกล้ๆกัน
a[2:-3]
```

```python
#เลือกเอาแค่ทั้งตำเเหน่ง -5 (negative number) ซึ่งนับมาจากข้างหลังจาก Array
a[-5:]
```
